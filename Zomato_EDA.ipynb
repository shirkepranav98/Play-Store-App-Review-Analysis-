```python
final_df= pd.merge(df,df2, on='Country Code', how='left')
final_df.head()
```

```python
##To check Data Types
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
Country_names = final_df.Country.value_counts().index
```

```python
Country_values = final_df.Country.value_counts().values
```

```python
# Pie chart---- Top 4 countries using zomato 
fig7, ax = plt.subplots(figsize = (10,8))
explode = (0.1,0.1,0.1,0,0.1)
plt.pie(Country_values[:4], labels = Country_names[:4],autopct='%1.3f%%')
```

Observation:Zomato maximum records or transaction are from India After that USA and then United Kingdom followed by Brazil.
