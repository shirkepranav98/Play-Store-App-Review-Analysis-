```python
#importing basics libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
test_df = pd.read_excel('/content/drive/MyDrive/AlmaBetter/Project on Flight Prediction/CSV Files/Test_set.xlsx')
test_df.head()
```

```python
final_df = train_df.append(test_df)
```

```python
# Feature Engg. Process
final_df['Date']= final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']= final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']= final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Date'] = final_df['Date'].astype(int)
```

```python
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
# Dropping the Date of Journey
final_df.drop('Date_of_Journey',axis=1)
```

```python
# Feature engg. on Arrival time
final_df.Arrival_Time = final_df.Arrival_Time.str.split(' ').str[0] 
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
# checking null values
final_df.isnull().sum()
```

```python
# Adding arrival hour and arrival minute
final_df['Arrival_hour']= final_df.Arrival_Time.str.split(':').str[0]
final_df['Arrival_min']= final_df.Arrival_Time.str.split(':').str[1]
```

```python
# converting above two in integer type
final_df['Arrival_hour'] = final_df.Arrival_hour.astype(int)
final_df['Arrival_min'] = final_df.Arrival_min.astype(int)
```

```python
# dropping the Arrival Time column
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
# Adding Dep hour and Dep minute
final_df['Dep_hour']= final_df.Dep_Time.str.split(':').str[0]
final_df['Dep_min']= final_df.Dep_Time.str.split(':').str[1]
```

```python
# converting above two in integer type
final_df['Dep_hour']= final_df.Dep_hour.astype(int)
final_df['Dep_min']= final_df.Dep_min.astype(int)
```

```python
# dropping the Dep time column
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df.Total_Stops.unique()
```

```python
# # # Mapping total_stops values
final_df.Total_Stops = final_df.Total_Stops.map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})
```
